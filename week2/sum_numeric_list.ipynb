```python
def sum_numeric_list(data):
    total = 0
    # ここに処理を記述
    for item in data:
      try:
        total += item
      except TypeError:
        print(f"無効な値が含まれています: {item}")
    return total

test_list1 = [1, 2, 3, 4, 5]
test_list2 = [1, "hello", 3, "world", 5]
print(f"リスト1の合計: {sum_numeric_list(test_list1)}")
print(f"リスト2の合計: {sum_numeric_list(test_list2)}")
```
